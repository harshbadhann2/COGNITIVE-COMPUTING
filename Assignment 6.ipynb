Q1. Ask the user to enter a value (e.g., M) for a mathematical function.

Generate x values from -10 to 10 using np.linspace().

Compute y values for:
o y= M⋅x2
o y=M⋅sin(x)

Plot both functions in a single figure:

o Use different colors and line styles.

o Add legend, grid, and title.

```python
import numpy as np
import matplotlib.pyplot as plt

N=float(input("Enter the value of N: "))
a=np.linspace(-15,15,300)
b1=N*a**2
b2=N*np.cos(a)

plt.plot(a, b1, 'g-', label=f'b = {N}a²')
plt.plot(a, b2, 'm--', label=f'b = {N}cos(a)')

plt.title(f'Mathematical Functions with N = {N}')
plt.xlabel('a')
plt.ylabel('b')
plt.grid(True)
plt.legend()
plt.show()
```

Q.2 Create a dataset of five subjects and scores.

Convert it into a Pandas DataFrame.  Plot the scores using a Seaborn bar plot with:

o Different colors for each bar.

o Annotations on top of each bar.

o Title, axis labels, and grid.

```python
import pandas as pd
import seaborn as sns
import random

info = {
'Topics': ['History', 'Geography', 'Biology', 'Art', 'Literature']}

df2 = pd.DataFrame(info)
df2['Grades'] = [random.randint(50, 90) for i in range(5)]

sns.barplot(x='Topics', y='Grades', data=df2, palette='coolwarm')

for i, grade in enumerate(df2['Grades']):
    plt.text(i, grade, str(grade), ha='center', va='bottom')

plt.title('Topic Grades Distribution')

plt.xlabel('Topics')

plt.ylabel('Grades')
plt.grid(True, axis='y')
plt.show()
```

Q3. Write a Python script to select your roll number as seed for NumPy and then generate a dataset of 50 values using np.random.randn().
Create a 2x2 subplot layout displaying:

o Line plot showing cumulative sum.

o Scatter plot with random noise.

Customize with titles, axis labels, and grids.

```python
import numpy as np
import matplotlib.pyplot as plt
student_id = 102317140
np.random.seed(student_id)
values = np.random.randn(50)

fig, ((plot1, plot2), (plot3, plot4)) = plt.subplots(2, 2, figsize=(12, 10))

cumulative_sum = np.cumsum(values)
plot1.plot(cumulative_sum, 'g-')
plot1.set_title('Cumulative Sum')
plot1.grid(True)
plot1.set_xlabel('Index')
plot1.set_ylabel('Cumulative Sum')

noise_values = np.random.normal(0, 0.01, 50)
plot2.scatter(range(50), values + noise_values, alpha=0.6)
plot2.set_title('Scatter Plot with Noise')
plot2.grid(True)
plot2.set_xlabel('Index')
plot2.set_ylabel('Value')

plot3.remove()
plot4.remove()

plt.show()
```
